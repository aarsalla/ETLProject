```python
import pandas as pd
import datetime as dr
```

```python
# The path to our CSV file
ks = "Resources/kickstarter.csv"
sp = "Resources/stocks.csv"

# Read our Kickstarter data into pandas
kickstarter_df = pd.read_csv(ks)

stocks_df = pd.read_csv(sp)
```

```python
# drop extraneous stock columns

stocks_df = stocks_df.drop(['High', 'Low', 'Adj Close', 'Volume'], axis=1)
stocks_df.head()
```

```python
kickstarter_df["launched"] = kickstarter_df["launched"].str.split(pat=" ", n=1, expand=True)
kickstarter_df.head()
```

```python
#2 smaller scripts that matches dates from two pandas datasets and appends stock data

launched_merge = pd.merge(kickstarter_df, stocks_df, how="left", left_on="launched", right_on="Date")
launched_merge.head(50)
```
